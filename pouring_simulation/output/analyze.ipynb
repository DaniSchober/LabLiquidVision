```python
import numpy as np
import matplotlib.pyplot as plt

TCP = np.loadtxt('CellFlask/Flask_35_200_22/TCP.txt', delimiter=',', skiprows=1)
theta_vs_volume = np.loadtxt('CellFlask/Flask_35_200_22/theta_vs_volume.txt', delimiter='\t', skiprows=1)
```

```python
# iterate over folders in CellFlask
import os

# Specify the path to the parent folder
parent_folder = 'CellFlask'

# Iterate over all entries in the parent folder
for entry in os.scandir(parent_folder):
    if entry.is_dir():
        folder_name = entry.name
        folder_path = entry.path + '/'

        # if folder name starts with 'Flask_45'
        if folder_name.startswith('Flask_35'):
            theta_vs_volume = np.loadtxt(folder_path + 'theta_vs_volume.txt', delimiter='\t', skiprows=1)
            plt.plot(theta_vs_volume[:,2], theta_vs_volume[:,0]/400)


        # plot the first column of theta_vs_volume against the second column
        #plt.plot(theta_vs_volume[:,2], theta_vs_volume[:,0]/300)

# axis labels
plt.ylabel('Remaining Volume (mL)')
plt.xlabel('Time(s)')
plt.title('Time vs Volume for Flask with 35mL starting volume')
# make plot bigger
plt.rcParams["figure.figsize"] = (30,10)
plt.show()
```

```python
# load the data from summary_flask.csv and skip rows 0 and 1 using pandas library
import pandas as pd
summary_flask = pd.read_csv('summary_flask.csv', skiprows=[0,1])


# plot the 6th column of summary_flask against the 8th column as points
#plt.plot(summary_flask.iloc[:,7], summary_flask.iloc[:,5], 'o')

# make colors of point based on the 7th column
plt.scatter(summary_flask.iloc[:,7], summary_flask.iloc[:,5], c=summary_flask.iloc[:,3])

# axis labels
plt.ylabel('Starting Volume (mL)')
plt.xlabel('Received Volume (mL)')
plt.title('Received Volume vs Starting Volume for Cell Culture Flask')

# explain colorbar
plt.colorbar(label='Maximum Theta (degrees)')

# make plot bigger
plt.rcParams["figure.figsize"] = (20,3)
plt.show()
```
