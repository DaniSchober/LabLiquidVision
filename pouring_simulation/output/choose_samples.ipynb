```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import pandas as pd

# Set font and fontsize globally
matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['font.size'] = 11
```

```python
# load the data from summary_flask.csv and skip rows 0 and 1 using pandas library
summary_flask = pd.read_csv('summary_flask.csv', skiprows=[0,1], )

# select randomly 50 rows from the summary_flask dataframe
summary_flask = summary_flask.sample(n=50)

# print the summary_flask dataframe
for i in range(len(summary_flask)):
    if summary_flask.iloc[i,6]*1.1 > summary_flask.iloc[i,5]:
        print("Sample folder:", summary_flask.iloc[i,1], ", Starting volume:", summary_flask.iloc[i,5], ", Poured volume:", summary_flask.iloc[i,5], ", Stop angle:", summary_flask.iloc[i,3], ", Stop time:", summary_flask.iloc[i,4], "\n")
        # write in text file
        with open("flask_test_samples.txt", "a") as f:
            f.write("Sample folder: " + str(summary_flask.iloc[i,1]) + ", Starting volume: " + str(summary_flask.iloc[i,5]) + ", Poured volume: " + str(summary_flask.iloc[i,5]) + ", Stop angle: " + str(summary_flask.iloc[i,3]) + ", Stop time: " + str(summary_flask.iloc[i,4]) + "\n\n")
    else:
        print("Sample folder:", summary_flask.iloc[i,1], ", Starting volume:", summary_flask.iloc[i,5], ", Poured volume:", summary_flask.iloc[i,6]*1.1, ", Stop angle:", summary_flask.iloc[i,3], ", Stop time:", summary_flask.iloc[i,4], "\n")
        # write in text file
        with open("flask_test_samples.txt", "a") as f:
            f.write("Sample folder: " + str(summary_flask.iloc[i,1]) + ", Starting volume: " + str(summary_flask.iloc[i,5]) + ", Poured volume: " + str(summary_flask.iloc[i,6]*1.2) + ", Stop angle: " + str(summary_flask.iloc[i,3]) + ", Stop time: " + str(summary_flask.iloc[i,4]) + "\n\n")

```
