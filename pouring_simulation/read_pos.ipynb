```python
import numpy as np

data = np.loadtxt('output/Assembly1_TCP.txt', delimiter=',', skiprows=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(data[:,0], label='pos_x, pos_y')
#plt.plot(data[:,1], label='pos_y')
#plt.plot(data[:,0], data[:,2], label='theta')
plt.xlabel('pos_x')
plt.ylabel('pos_y')
plt.legend()
plt.show()
```

```python
#plt.plot(data[:,0], label='pos_x, pos_y')
plt.plot(data[:,1], label='pos_y')
#plt.plot(data[:,0], data[:,2], label='theta')
plt.xlabel('pos_x')
plt.ylabel('pos_y')
plt.legend()
plt.show()
```
